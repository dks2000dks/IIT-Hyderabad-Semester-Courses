## Importing Modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

## *Functions*

```python
def Sin_Continous(a,b,A = 1,f = 1):										# Function to generate and plot a Continous Sin
	"""
	A = Amplitude
	f = Frequency
	"a" and "b" are range for plotting the graph
	"""
	time = np.arange(a,b,0.01)
	amplitude = A*np.sin(f*time)
	
	return amplitude,time
```

```python
def Cos_Continous(a,b,A = 1,f = 1):										# Function to generate and plot a Continous Cos
	"""
	A = Amplitude
	f = Frequency
	"a" and "b" are range for plotting the graph
	"""
	time = np.arange(a,b,0.01)
	amplitude = A*np.cos(f*time)
	
	return amplitude,time
```

```python
def Square_Wave(a,b,L=1500):											# Generating Square Wave
	"""
	L = Limit
	"""
	
	N = np.arange(1,L,2)												
	time = np.arange(a,b,0.01)
	s = time.shape[0]
	output = np.zeros(s)
	
	for n in N:
		output += (Sin_Continous(a,b,A=1/(n),f=n))[0]
		
	output = (output - min(output))/(max(output) - min(output))
	
	plt.xlabel("Time")
	plt.ylabel("Output")
	plt.plot(time,output)
	plt.show()
		
	return output,time
```

```python
def Square_Wave_Excat(m,n,T=1,s=10):											# Generating Square Wave
	"""
	Plotting Square between -mT to nT.
	T = Time Period
	s = Sampling
	"""
    
	N = np.arange(m,n+1,1)
	o = N.shape[0]
	y = [0]
	x = [0]
	
	for i in N:
		x1 = list(np.linspace(i*T,i*T + 0.5,s))
		y1 = list(np.ones(s))

		x2 = list(np.linspace(i*T+0.5,i*T +1,s))
		y2 = list(-1*np.ones(s))


		x = x + x1 + x2

		y = y + y1 + y2
		
	y.pop(0)
	x.pop(0)
	
	plt.xlabel("Time")
	plt.ylabel("Output")
	plt.plot(x,y)
	plt.show()
		
	return y,x
```

```python
def Triangle_Wave(a,b,L=1500):											# Generating Triangle Wave
	"""
	L = Limit
	"""
	
	N = np.arange(1,L,2)												
	time = np.arange(a,b,0.01)
	s = time.shape[0]
	output = np.zeros(s)
	
	for n in N:
		output += (Cos_Continous(a,b,A=1/(n**2),f=n))[0]
	
	output = (output - min(output))/(max(output) - min(output))
		
	plt.xlabel("Time")
	plt.ylabel("Output")
	plt.plot(time,output)
	plt.show()
		
	return output,time
```

```python
def Triangle_Wave_Excat(m,n,T=1,s=10):											# Generating Square Wave
	"""
	Plotting Square between -mT to nT.
	T = Time Period
	s = Sampling
	"""
    
	N = np.arange(m,n,1)
	o = N.shape[0]
	y = [0]
	x = [0]
	
	for i in N:
		x1 = list(np.linspace(i*T,i*T + 0.5,s))
		y1 = list(np.linspace(0.5,0,s))

		x2 = list(np.linspace(i*T+0.5,i*T +1,s))
		y2 = list(np.linspace(0,0.5,s))


		x = x + x1 + x2

		y = y + y1 + y2
		
	y.pop(0)
	x.pop(0)
	
	plt.xlabel("Time")
	plt.ylabel("Output")
	plt.plot(x,y)
	plt.show()
		
	return y,x
```

```python
def Fourier_Coeff_Square_Wave(L=1500):								# Generating Square Wave Fourier Spectrum
	"""
	L = Limit
	"""
	
	N = np.arange(1,L+1,2)
	u = N.shape[0]
	spectrum = np.zeros(u)

	for n in N:
		spectrum[int(n/2)] = 4/(np.pi*n)
		
	plt.stem(N,spectrum)
	plt.show()

	return N,spectrum
```

```python
def Fourier_Series_Coeff_Square_Wave(L=1500):								# Generating Square Wave Fourier Spectrum
	"""
	L = Limit
	"""
	
	k = np.arange(1,L+1,2)
	s = k.shape[0]
	N = np.zeros(2*s)
	N[0:s]= -k[::-1]
	N[s:2*s]= k
	i = 0
	
	u = N.shape[0]
	spectrum = np.zeros(u)
	phase = np.zeros(u)
		
	for n in N:
		spectrum[i] = abs(2/(np.pi*n))
		phase[i] = (n/abs(n))*(np.pi)*0.5
		
		i = i+1
	
	plt.figure(figsize=(13, 8))
	ax = plt.subplot(1,2,1)
	plt.stem(N,phase)
	
	ax = plt.subplot(1,2,2)
	plt.stem(N,spectrum)
	plt.show()

	return N,spectrum
```

```python
def Fourier_Coeff_Triangle_Wave(L=1500):							# Generating Triangle Wave Fourier Spectrum
	"""
	L = Limit
	"""
	
	N = np.arange(1,L+1,2)
	u = N.shape[0]
	spectrum = np.zeros(u)
	
	for n in N:
		spectrum[int(n/2)] = 2/(np.pi*n)**2
	
	plt.stem(N,spectrum)
	
	plt.show()
		
	return N,spectrum
```

```python
def Fourier_Series_Coeff_Triangle_Wave(L=1500):							# Generating Triangle Wave Fourier Spectrum
	"""
	L = Limit
	"""
	
	k = np.arange(1,L+1,2)
	s = k.shape[0]
	N = np.zeros(2*s)
	N[0:s]= -k[::-1]
	N[s:2*s]= k
	i = 0
	
	u = N.shape[0]
	spectrum = np.zeros(u)
	phase = np.zeros(u)
		
	for n in N:
		spectrum[i] = abs(1/(2*np.pi*n**2))
		phase[i] = (n**2/abs(n**2))*(np.pi)*0.5
		
		i = i+1
	
	plt.figure(figsize=(13, 8))
	ax = plt.subplot(1,2,1)
	plt.stem(N,phase)
	
	ax = plt.subplot(1,2,2)
	plt.stem(N,spectrum)
	plt.show()

	return N,spectrum
```

### 1e)

*Let D be the desired Duty Cycle. By Reverse-Engineering*

```python
def function1(D,A,a=0,b=4*np.pi,L=5000):								
	"""
	L = Limit
	"""
	
	N = np.arange(1,L,1)
	time = np.arange(a,b,0.01)
	s = time.shape[0]
	output = np.zeros(s)
	
	for n in N:
		p = (50/(np.pi*n*D))
		output += (Sin_Continous(a,b,A=p,f=n))[0]
	
	output = A*(output * 2	+ 1 - (50/D))											# Symmetric.So Negetives and Positives combine to form 2 times summation.
	
	output = A * (output)/(max(output) - min(output))

	plt.plot(time,output)
	plt.show()
		
	return output,time
	
def function2(D,A,a=0,b=4*np.pi,L=5000):							
	"""
	L = Limit
	"""
	
	N = np.arange(1,L,1)
	time = np.arange(a,b,0.01)
	s = time.shape[0]
	output = np.zeros(s)
	
	for n in N:
		p = (np.sin(n*np.pi*D/50))*(1/(n*np.pi))
		q = (1 - np.cos(n*np.pi*D/50)) * (1/(n*np.pi))
		output += (Sin_Continous(a,b,A=q,f=n))[0]
		output += (Cos_Continous(a,b,A=p,f=n))[0]
	
	output = output * 2															# Symmetric.So Negetives and Positives combine to form 2 times summation.
	
	output = A * (output)/(max(output) - min(output))

	plt.plot(time,output)
	plt.show()
		
	return output,time
	

# For 10% Duty Cycle

dummy1,dummy2 = function1(10,1)      
dummy1,dummy2 = function2(10,1)
```
