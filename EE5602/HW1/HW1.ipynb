## Disparity Estimation for Stereo Image Pair

**Assumptions:**
- Left Image is considered as Reference.
- $3 \times 3$ Window is considered for each Candidate Disparity to increase robustness of Matching Process.


**Sum-Product Algorithm:**

\begin{align}
\mu_{s}\left(x_{s}\right):=\sum_{\left\{x^{\prime} \mid x_{s}^{\prime}=x_{s}\right\}} p\left(x_{1}^{\prime}, x_{2}^{\prime}, \ldots, x_{m}^{\prime}\right) = \kappa \psi_{s}\left(x_{s}\right) \prod_{t \in N(s)} M_{t s}^{*}\left(x_{s}\right) \\
M_{t s}^{*}\left(x_{s}\right):= \sum_{x_{V_{t}}^{\prime}} \psi_{s t}\left(x_{s}, x_{t}^{\prime}\right) p\left(x_{V_{t}}^{\prime} ; T_{t}\right) \\
p\left(x_{V_{t}} ; T_{t}\right) \propto \prod_{u \in V_{t}} \psi_{u}\left(x_{u}\right) \prod_{(u, v) \in E_{t}} \psi_{u v}\left(x_{u}, x_{v}\right) \\
\psi_{S}\left(d_{s}\right) \propto \exp \left\{-\frac{1}{2 \sigma^{2}}\right.\left.\left(I_{L}\left(i_{s}, j_{s}\right)-I_{R}\left(i_{s}, j_{s}+d_{s}\right)\right)^{2}\right\} \\
\psi_{s t}\left(d_{s}, d_{t}\right) \propto \exp \left\{-\frac{1}{2 \gamma^{2}} \min \left(\left(d_{s}-d_{t}\right)^{2}, \delta_{0}^{2}\right)\right\}
\end{align}


---

**Parameters:**
1. **K**: No.of Levels of Random Variable $d_{s}$.
2. $\sigma$ **(Sigma)**: Standard Deviation of Gaussian Compatability Function.
3. $\gamma$ **(Gamma)**: Standard Deviation of Truncated Gaussian Compatability Function between 2 Nodes.
4. $\delta_{0}$ **(Delta)**: Threshold for the Maximum Degree to Large Disparities.
5. $\mu_{s}$: Marginal Distribution.

```python
class DisparityEstimation:
    def __init__(self,K):
        # Initialisations
        self.K = K
        self.N = None
        self.M = None
        
    def Compatibility(self,i,j):
        # Gaussian Compatibility Function
        return np.exp((-0.5/pow(self.Sigma,2)) * np.square(self.IL[i,j] - self.IR[i,j:min(self.M, j+self.K)]))
    
    def CompatibilityNeighbourHood(self):
        # Estimating Truncated Gaussian Compatibility between 2 Nodes
        self.CN = np.zeros((self.K,self.K))
        for ds in range(self.K):
            for dt in range(self.K):
                self.CN[ds][dt] = np.exp((-0.5/pow(self.Gamma,2)) * min(np.square(ds-dt),pow(self.Delta,2)))
    
    def NeighbourhoodMessages(self,i,j):
        # Messages from Neigbouring Modes
        Message = np.ones((self.K,))
        for x in range(-1,2):
            for y in range(-1,2):
                if (x!=0 and y!=0 and (i+x)<self.N and (j+y)< self.M and (i+x)>=0 and (j+y)>=0):
                    Data = np.zeros((self.K,))
                    for s in range(self.K):
                        Data[s] = np.dot(self.CN[s].T,self.CM[i+x,j+y])
                    Message = np.multiply(Message,Data)
        return Message
                    
    
    def Disparity(self):
        # Estimating Disparity of each Pixel
        DisparityMatrix = np.zeros((self.N,self.M))
        for i in range(self.N):
            for j in range(self.M):
                Data = np.multiply(self.CM[i][j], self.NeighbourhoodMessages(i,j))
                DisparityMatrix[i][j] = np.argmax(Data)
                
        return DisparityMatrix
                
        
    def EstimateDisparity(self, Left, Right, Sigma=1, Gamma=1, Delta=5):
        # Estimating Disparity between Images
        self.IL = Left
        self.IR = Right
        self.N, self.M = self.IL.shape
        self.Sigma = Sigma
        self.Gamma = Gamma
        self.Delta = Delta
        
        self.CM = np.zeros((self.N,self.M,self.K))
        
        # Estimating Compatibility Matrix
        for i in range(self.N):
            for j in range(self.M):
                s = self.Compatibility(i,j)
                self.CM[i,j,0:s.shape[0]] = self.Compatibility(i,j)
                
        #Estimating Truncated Gaussian Compatibility
        self.CompatibilityNeighbourHood()
        
        # Calcula Disparity Map
        self.DisparityMatrix = self.Disparity()
        
    def PlotDisparity(self):
        # Plotting Disparity Map
        plt.figure(figsize=(25,15))
        
        plt.subplot(1,3,1)
        plt.title("Left Image")
        plt.imshow(Left,cmap='gray')

        plt.subplot(1,3,2)
        plt.title("Right Image")
        plt.imshow(Right,cmap='gray')
        
        plt.subplot(1,3,3)
        plt.title("Disparity Map")
        plt.imshow(self.DisparityMatrix,cmap='gray',vmin=0,vmax=self.K-1)

        plt.show()
```

### Disparity Map for Stereo Image Pair

- $\delta_{0} = 5$

```python
DE = DisparityEstimation(10)
DE.EstimateDisparity(Left,Right)
```
